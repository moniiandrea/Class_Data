```python
# Text Classification

import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
# Training data
documents = [
    ("Your flight reservation with American Airlines", "Real"),
    ("Invoice for your Order #123456", "Real"),
    ("Meeting scheduled – Zoom link attached", "Real"),
    ("Welcome to Azure – Access information", "Real"),
    ("Your monthly bank statement is now available", "Real"),
    ("Congratulations! You've won an iPhone 15! Claim your prize now.", "Spam"),
    ("URGENT: Your bank account has been locked. Click here to verify.", "Spam"),
    ("Work from home and earn $5,000 a month with no experience.", "Spam"),
    ("Dear friend, I need your help transferring $10 million", "Spam"),
    ("Discover the secret to losing 10 kg in a week without exercise!", "Spam")

]

```

```python
# Prepare features and labels
vectorizer = CountVectorizer()
features = vectorizer.fit_transform([doc[0] for doc in documents])
labels = [doc[1] for doc in documents]

# Train a classifier (Naive Bayes)
classifier = MultinomialNB()
classifier.fit(features, labels)
```

```python
# Test with a new example
# 1. "Security update: Please reset your password" - Real
# 2. "Congratulations! Exclusive offer: medications at 90% discount!" - Spam
new_example = vectorizer.transform(["Congratulations! Exclusive offer: medications at 90% discount!"])
prediction = classifier.predict(new_example)
print(prediction)
```
